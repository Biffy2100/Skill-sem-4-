```python
# prompt: comprehensive data visualisation using iris dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['target_name'] = iris.target_names[iris.target]

# 1. Pairplot to visualize relationships between all features
sns.pairplot(iris_df, hue='target_name', diag_kind='kde')
plt.show()

# 2. Boxplots to compare feature distributions across different species
plt.figure(figsize=(12, 6))
for i, feature in enumerate(iris.feature_names):
  plt.subplot(2, 2, i+1)
  sns.boxplot(x='target_name', y=feature, data=iris_df)
  plt.title(feature)
plt.tight_layout()
plt.show()

# 3. Violin plots to visualize feature distributions and density across different species
plt.figure(figsize=(12, 6))
for i, feature in enumerate(iris.feature_names):
  plt.subplot(2, 2, i+1)
  sns.violinplot(x='target_name', y=feature, data=iris_df)
  plt.title(feature)
plt.tight_layout()
plt.show()

# 4. Scatter plot to visualize the relationship between two features and color by species
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target_name', data=iris_df)
plt.title('Sepal Length vs Sepal Width')
plt.show()

# 5. 3D scatter plot to visualize the relationship between three features and color by species
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
colors = ['red', 'green', 'blue']
for target, color in zip(iris_df['target'].unique(), colors):
  ax.scatter(iris_df.loc[iris_df['target'] == target, 'sepal length (cm)'],
             iris_df.loc[iris_df['target'] == target, 'sepal width (cm)'],
             iris_df.loc[iris_df['target'] == target, 'petal length (cm)'],
             c=color, label=iris.target_names[target])
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
ax.legend()
plt.show()

# 6. Correlation matrix heatmap to visualize feature correlations
plt.figure(figsize=(8, 6))
sns.heatmap(iris_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 7. Histogram to visualize the distribution of a specific feature
plt.figure(figsize=(8, 6))
sns.histplot(iris_df['sepal length (cm)'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()

# 8. KDE plot to visualize the density distribution of a specific feature
plt.figure(figsize=(8, 6))
sns.kdeplot(iris_df['sepal length (cm)'])
plt.title('Density Distribution of Sepal Length')
plt.show()
```
