```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['target_name'] = iris.target_names[iris.target]

# Set Seaborn style for better aesthetics
sns.set_style("whitegrid")

# 1. Pairplot to visualize relationships between all features
sns.pairplot(iris_df, hue='target_name', diag_kind='kde', height=2.5, palette='Set1')
plt.show()

# 2. Boxplots with Swarmplot Overlay
plt.figure(figsize=(12, 6))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='target_name', y=feature, data=iris_df)
    sns.swarmplot(x='target_name', y=feature, data=iris_df, color=".25", alpha=0.7)
    plt.title(feature)
plt.tight_layout()
plt.show()

# 3. Violin plots to visualize feature distributions and density
plt.figure(figsize=(12, 6))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    sns.violinplot(x='target_name', y=feature, data=iris_df, palette='muted')
    plt.title(feature)
plt.tight_layout()
plt.show()

# 4. Scatter plot with improved visibility
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target_name', data=iris_df, alpha=0.7)
plt.title('Sepal Length vs Sepal Width')
plt.show()

# 5. 3D scatter plot for better feature visualization
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
colors = ['red', 'green', 'blue']
for target, color in zip(iris_df['target'].unique(), colors):
    ax.scatter(iris_df.loc[iris_df['target'] == target, 'sepal length (cm)'],
               iris_df.loc[iris_df['target'] == target, 'sepal width (cm)'],
               iris_df.loc[iris_df['target'] == target, 'petal length (cm)'],
               c=color, label=iris.target_names[target], alpha=0.7)
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
ax.legend()
plt.show()

# 6. Correlation heatmap with better formatting
plt.figure(figsize=(8, 6))
sns.heatmap(iris_df.drop(columns=['target_name']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()


# 7. Histogram with KDE
plt.figure(figsize=(8, 6))
sns.histplot(iris_df['sepal length (cm)'], kde=True, bins=15, color='royalblue')
plt.title('Distribution of Sepal Length')
plt.show()

# 8. KDE plot for density distribution
plt.figure(figsize=(8, 6))
sns.kdeplot(iris_df['sepal length (cm)'], fill=True, color='purple', alpha=0.7)
plt.title('Density Distribution of Sepal Length')
plt.show()
```
